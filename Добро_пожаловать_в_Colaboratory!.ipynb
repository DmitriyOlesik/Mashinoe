```
df = pd.read_excel('копия.xlsx')
df
```

```
df = df[['Адресат',
         'Автор',
         'Краткое\nсодержание',]]
df
```

```
old_column = df['Автор']
authors = list()
organisations = list()
for data in old_column:
  author = data[:data.find('(') - 1]
  organisation = data[data.find('(') +1:-1]
  authors.append(author)
  organisations.append(organisation)

df = df.assign(author_name=authors)
df = df.assign(organisation_name=organisations)
df = df.drop(columns=['Автор'])
df
```
