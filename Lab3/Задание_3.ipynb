````python
import pandas as pd
import numpy as np
def create_holes(dataframe, column_name, hole_percentage = 10):

    if hole_percentage < 0 or hole_percentage > 100:
        raise ValueError("Процент дырявости должен быть в диапазоне от 0 до 100.")

    dataframe_with_holes = dataframe.copy()
    num_holes = int(len(dataframe) * (hole_percentage / 100))
    hole_indices = np.random.choice(len(dataframe), num_holes, replace=False)
    dataframe_with_holes.loc[hole_indices, column_name] = np.nan

    return dataframe_with_holes


df = create_holes(df, 'Rape')
df
````

Данный код создает функцию createholes, которая принимает в качестве параметров исходный датафрейм, имя столбца и процент "дырявости" (процент пропущенных значений), по умолчанию равный 10%.

Функция создает копию исходного датафрейма и затем определяет количество "дыр" (пропущенных значений) на основе заданного процента. Затем случайным образом выбирает индексы строк, которые будут иметь пропущенные значения в указанном столбце, используя функцию np.random.choice().

После этого функция устанавливает выбранные значения в столбце равными np.nan, что представляет собой пропущенное значение в pandas.

Таким образом, результатом выполнения функции createholes будет исходный датафрейм с определенным процентом строк в указанном столбце, содержащими пропущенные значения (NaN).

Дубликаты

Проверьте датафрейм на дубликаты и удалите их, если дублей нет - создайте.

````python
#Проверка на дубли
df = pd.read_csv('kc_house_data.csv')
dupls = df.duplicated()

count = 0
for i in dupls:
  if (i): count +=1

count
````

Этот код загружает CSV-файл в датафрейм df, затем использует метод df.duplicated(), чтобы найти дубликаты в датафрейме.

Затем, с помощью цикла for, проходим по каждому значению в массиве dupls (который возвращает True для каждого дубликата и False для каждого уникального значения) и увеличиваем переменную count на 1 каждый раз, когда встречаем дубликат.

Конечное значение переменной count будет представлять собой количество дубликатов в датафрейме.


````python
df_with_dupls = pd.concat((df, df.loc[0:100]))
dupls = df_with_dupls.duplicated()
count = 0
for i in dupls:
  if (i): count +=1

count
````

Этот код создает новый датафрейм dfwithdupls, путем конкатенации (объединения) исходного датафрейма df с собой же (df.loc0:100). После этого dfwithdupls содержит исходные строки из df, за которыми следует их первые 100 строк.

Затем для нового датафрейма dfwithdupls используется метод df.duplicated(), чтобы найти дубликаты.

Затем, с помощью цикла for, мы проходим по каждому значению в массиве dupls и увеличиваем переменную count на 1 каждый раз, когда встречаем дубликат.

Конечное значение переменной count будет представлять собой количество дубликатов в датафрейме dfwithdupls, включая дубликаты, возникшие из-за конкатенации исходного датафрейма.


````python
#Удаление дублей
df_no_duplicates = df_with_dupls.drop_duplicates()
dupls = df_no_duplicates.duplicated()

count = 0
for i in dupls:
  if (i): count +=1
count
````

Этот код удаляет дубликаты из датафрейма dfwithdupls с помощью метода dropduplicates() и сохраняет изменения в новом датафрейме dfnoduplicates.

Затем используется метод df.duplicated() для нового датафрейма dfnoduplicates для поиска дубликатов.

Затем, с помощью цикла for, проходим по каждому значению в массиве dupls и увеличиваем переменную count на 1 каждый раз, когда встречаем дубликат.

Конечное значение переменной count будет представлять собой количество дубликатов в датафрейме dfnoduplicates. Однако, после удаления дубликатов с использованием метода dropduplicates, все значения в массиве dupls будут равны False, так как датафрейм dfnoduplicates уже не содержит дубликатов, и поэтому переменная count будет равна 0.

3. С помощью метода fittransform() объекта scalerminmax, значения числовых признаков dfnum_features масштабируются с использованием метода минимакс-шкалирования (MinMaxScaler) и затем данные присваиваются копии датафрейма minmaxscaler.

Данный код позволяет масштабировать числовые признаки с сохранением пропорций между значениями в диапазоне от 0 до 1, что может улучшить процесс обучения модели, особенно при использовании алгоритмов, чувствительных к масштабу.
